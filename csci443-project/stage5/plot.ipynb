```python
import matplotlib.pyplot as plt
import numpy as np
```

## Task Performance Computation

```python

# The following data was taken from the videos of them completing the tasks.
task_secs = np.array([
    [39-17, 7, 10, 110, 5],
    [40, 32, 28, 27, 20],
    [40, 8, 33, 40, 6],
    [20, 5, 21, 60, 3]
])

task_clicks = np.array([
    [2, 2, 3, 18, 2],
    [3, 2, 3, 8, 2],
    [2, 2, 3, 4, 2],
    [2, 2, 3, 8, 2]
])

num_tasks = 5

# Plot Averages
plt.bar([f"Task {i + 1}" for i in range(num_tasks)], [np.mean(task_secs[:, i]) for i in range(num_tasks)])
plt.xlabel('Tasks')
plt.ylabel('Average Seconds')
plt.title('Average Seconds per Task')
plt.show()

plt.bar([f"Task {i + 1}" for i in range(num_tasks)], [np.mean(task_clicks[:, i]) for i in range(num_tasks)])
plt.xlabel('Tasks')
plt.ylabel('Average Clicks')
plt.title('Average Clicks per Task')
plt.show()

# Plot STDs
plt.bar([f"Task {i + 1}" for i in range(num_tasks)], [np.std(task_secs[:, i]) for i in range(num_tasks)])
plt.xlabel('Tasks')
plt.ylabel('STDs')
plt.title('Standard Deviation of Seconds per Task')
plt.show()

plt.bar([f"Task {i + 1}" for i in range(num_tasks)], [np.std(task_clicks[:, i]) for i in range(num_tasks)])
plt.xlabel('Tasks')
plt.ylabel('STDs')
plt.title('Standard Deviation of Clicks per Task')
plt.show()
```

## SUS Score Computation
The following code computes the SUS scores based on the google form they filled out after completing the tasks.

```python
# Scores copy pasted from google forms
sus = np.array([
    [3,	3,	3,	2,	4,	3,	4,	2,	2,	2],
    [2,	3,	3,	1,	3,	4,	4,	1,	2,	2],
    [3,	2,	3,	2,	4,	1,	4,	2,	3,	2],
    [4,	2,	3,	1,	4,	3,	4,	2,	4,	3],
])

# Computes the SUS score from the provided ratings
# This follows the formula listed on wikipedia: https://en.wikipedia.org/wiki/System_usability_scale
def computeSUSScore(row):
   oddSum = sum([row[idx]-1 for idx in range(len(row)) if (idx + 1) % 2 != 0])
   evenSum = sum([row[idx]-1 for idx in range(len(row)) if (idx + 1) % 2 == 0])

   return 2.5 * (20 + oddSum - evenSum)

scores = [computeSUSScore(row) for row in sus]

print(f"The average SUS score is {np.mean(scores)}")
print(f"The Standard Deviation of the SUS score's is {np.std(scores)}")

# plot the scores
plt.bar([f"Person {i}" for i in range(len(sus))], scores)
plt.axhline(68, color='red', label='Baseline SUS Score (68)')

plt.xlabel('Participant')
plt.ylabel('SUS Score')
plt.title('SUS Scores by Participant')
plt.ylim(0, 100) # show entire range
plt.legend()
plt.show()
```
